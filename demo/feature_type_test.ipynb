```python
import pandas as pd
import numpy as np
from sklearn.pipeline import make_pipeline
from skrebate import ReliefF
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

### Example dataset description

The dataset being analyzed is a synthetic genetic dataset generated by GAMETES, which simulates genetic interactions (epistasis). It comprises 1600 samples, each with 18 non-informative (low importance) features labeled as N0 through N17, and 2 informative (high importance) features labeled as P1 and P2. The final column, labeled class, indicates the binary classification outcome.

```python
genetic_data = pd.read_csv('./data/GAMETES_Epistasis_2-Way_20atts_0.4H_EDM-1_1.tsv.gz',
                           sep='\t', compression='gzip')
print(f'Total data size: {len(genetic_data)}')
genetic_data.head()
```

```python
features, labels = genetic_data.drop('class', axis=1).values, genetic_data['class'].values
```

## My test

```python
from sklearn.model_selection import train_test_split
```

```python
# Didn't specify anything - all continuous

X_train, X_test, y_train, y_test = train_test_split(features, labels)
print(f'Training data: {len(X_train)}; Testing data: {len(X_test)}')

fs = ReliefF()
fs.fit(X_train, y_train)

summary = fs.summary()

for feature_name, data_type, feature_score in zip(genetic_data.drop('class', axis=1).columns, summary['data type'],
                                       fs.feature_importances_):
    print(feature_name, '\t', data_type, '\t', feature_score)
```

```python
# Specify the 1st and 2nd as categorical features

fs = ReliefF(categorical_features=[0, 1])
fs.fit(X_train, y_train)

summary = fs.summary()

for feature_name, data_type, feature_score in zip(genetic_data.drop('class', axis=1).columns, summary['data type'],
                                       fs.feature_importances_):
    print(feature_name, '\t', data_type, '\t', feature_score)
```

```python
# Choose all as categorical features

fs = ReliefF(categorical_features=list(range(X_train.shape[1])))
fs.fit(X_train, y_train)

summary = fs.summary()

for feature_name, data_type, feature_score in zip(genetic_data.drop('class', axis=1).columns, summary['data type'],
                                       fs.feature_importances_):
    print(feature_name, '\t', data_type, '\t', feature_score)
```
