```python
import pandas as pd
import numpy as np
from sklearn.pipeline import make_pipeline
from skrebate import ReliefF
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
genetic_data = pd.read_csv('./data/GAMETES_Epistasis_2-Way_continuous_endpoint_a_20s_1600her_0.4__maf_0.2_EDM-2_01.tsv.gz',
                           sep='\t', compression='gzip')
print(f'Total data size: {len(genetic_data)}')
genetic_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features, labels = genetic_data.drop('Class', axis=1).values, genetic_data['Class'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels)
print(f'Training data: {len(X_train)}; Testing data: {len(X_test)}')

fs = ReliefF(n_features_to_select=2, n_neighbors=100)
fs.fit(X_train, y_train)

for feature_name, feature_score in zip(genetic_data.drop('Class', axis=1).columns,
                                       fs.feature_importances_):
    print(feature_name, '\t', feature_score)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels)
print(f'Training data: {len(X_train)}; Testing data: {len(X_test)}')

fs = ReliefF(n_features_to_select=2, n_neighbors=100)
fs.fit(X_train, y_train)

summary = fs.summary()

for feature_name, data_type, feature_score in zip(genetic_data.drop('Class', axis=1).columns, summary['data type'],
                                       fs.feature_importances_):
    print(feature_name, '\t', data_type, '\t', feature_score)
```
